```python
s = pyperclip.paste()
print(s)
```

```python
regex = re.compile(r'\t(.{2,5})\r\n')
numbers = regex.findall(s)[1:]
```

```python
numbers
len(numbers)
print('\t'.join(numbers[:5]))
```

```python
with open('data.txt', 'a') as f:
    f.write('\t'.join(numbers))
```
